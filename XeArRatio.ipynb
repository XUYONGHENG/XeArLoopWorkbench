```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt


xenonHeatmap = pd.read_csv("Xenon_Signal_Significance_Heatmap_zoomedIn.csv", index_col=0)
argonHeatmap = pd.read_csv("Argon_Signal_Significance_Heatmap_zoomedIn.csv", index_col=0)

xenonHeatmap = xenonHeatmap.iloc[::]
argonHeatmap = argonHeatmap.iloc[::]
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(xenonHeatmap, 
            annot=False, 
            fmt=".3e", 
            cmap="rocket_r", 
            xticklabels=5,
            yticklabels=5,
            cbar=True)  # cbar=True 添加颜色条
plt.title("Heatmap with Colorbar")
plt.xlabel("Column Header (mchi)")
plt.ylabel("Row Index (gchi)")
plt.show()
```

```python
difference = xenonHeatmap - argonHeatmap

fig, ax = plt.subplots(figsize=(6, 4.5),dpi=200)
sns.heatmap(difference, 
            annot=False, 
            fmt=".3e", 
            cmap="rocket_r", 
            xticklabels=5,
            yticklabels=5,
            vmin=-1,
            vmax=7,
            cbar=True)  # cbar=True 添加颜色条
ax.set_title(r"Ratio between Xenon and Argon significances $\log_{10}\frac{\mathcal{S}_{Xe}}{\mathcal{S}_{Ar}}$")
ax.set_xlabel(r"$m_\chi$ [GeV]")
ax.set_ylabel(r"$g_\chi$")
fig.tight_layout()
fig.savefig("XeArRatio_zoomedIn.pdf",dpi=200)
```
