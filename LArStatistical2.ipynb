```python
DS_LM_exposure = 1 # tonne/year

def significance(signalRate, bkgRate, exposure):
    return signalRate * exposure / np.sqrt(bkgRate*exposure)

print(significance(totalSignal_ar50p, bkg_ref, DS_LM_exposure), significance(totalSignal_ar50pl, bkg_ref, DS_LM_exposure), significance(totalSignal_ar50z, bkg_ref, DS_LM_exposure), significance(totalSignal_ar50zl, bkg_ref, DS_LM_exposure))
print(significance(totalSignal_ar200p, bkg_ref, DS_LM_exposure), significance(totalSignal_ar200pl, bkg_ref, DS_LM_exposure), significance(totalSignal_ar200z, bkg_ref, DS_LM_exposure), significance(totalSignal_ar200zl, bkg_ref, DS_LM_exposure))

print("50 GeV PS-PS No Higgs", significance(totalSignal_ar50p, bkg_ref, DS_LM_exposure))
print("50 GeV PS-PS Higgs", significance(totalSignal_ar50pl, bkg_ref, DS_LM_exposure))
print("50 GeV S-PS No Higgs", significance(totalSignal_ar50z, bkg_ref, DS_LM_exposure))
print("50 GeV S-PS Higgs", significance(totalSignal_ar50zl, bkg_ref, DS_LM_exposure))
print("200 GeV PS-PS No Higgs", significance(totalSignal_ar200p, bkg_ref, DS_LM_exposure))
print("200 GeV PS-PS Higgs", significance(totalSignal_ar200pl, bkg_ref, DS_LM_exposure))
print("200 GeV S-PS No Higgs", significance(totalSignal_ar200z, bkg_ref, DS_LM_exposure))
print("200 GeV S-PS Higgs", significance(totalSignal_ar200zl, bkg_ref, DS_LM_exposure))

```
